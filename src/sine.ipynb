```python
# The sine function along y-axis

amplitude = 1.0
frequency = 3.0
period = 1 / frequency
samples = np.linspace(0, np.pi/2, 100)

y = amplitude * np.cos(2*np.pi*frequency*period*samples)

fig = plt.figure(figsize=(2, 6))
ax = fig.add_subplot(111)
ax.plot(y, samples)
ax.set_ylim(samples.min(), samples.max())
ax.set_xlim(np.pi/2, -np.pi/2)
# ax.set_axis_off()
None
```
