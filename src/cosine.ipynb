```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The cosine function along x-axis

amplitude = 1.0
frequency = 3.0
period = 1 / frequency
samples = np.linspace(0, np.pi/2, 100)

y = amplitude * np.cos(2*np.pi*frequency*period*samples)

fig = plt.figure(figsize=(6, 2))
ax = fig.add_subplot(111)
ax.plot(samples, y)
ax.set_xlim(0, np.pi/2)
ax.set_ylim(-np.pi/2, np.pi/2)
# ax.set_axis_off()
None
```
